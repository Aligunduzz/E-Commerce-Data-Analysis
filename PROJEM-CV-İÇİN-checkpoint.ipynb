order_delivered_customer_date ve order_delivered_carrier_date sütunlarındaki eksik değerler analiz edildi.
Eksik değerlerin büyük çoğunluğu henüz tamamlanmamış (processing, invoiced, approved, shipped) veya iptal edilmiş (canceled, unavailable) siparişlerden kaynaklanmaktadır.
Bu nedenle bu kayıtlar veri setinden çıkarılmadı.
Ancak order_status = "delivered" olup teslim veya kargoya veriliş tarihi eksik olan çok az sayıda kayıt (8 ve 2 adet) veri tutarsızlığı olarak not edildi.

```python
delivered_drop = (
    orders["order_delivered_customer_date"].isnull() &
    orders["order_delivered_carrier_date"].isnull() &
    (orders["order_status"] == "delivered")
)
```

```python
orders = orders.drop(orders[delivered_drop].index)
```

```python
orders.loc[mask2,"order_status"].value_counts()
```

```python
delivered_drop2 = (
    orders["order_delivered_customer_date"].isnull() &
    (orders["order_status"] == "delivered")
)
```

```python
orders = orders.drop(orders[delivered_drop2].index)
```

```python
delivered_drop3 = (
    orders["order_delivered_carrier_date"].isnull() &
    (orders["order_status"] == "delivered")
)
```

```python
orders = orders.drop(orders[delivered_drop3].index)
```

Order_Delivered_Customer_Date Eksik değer analizi sonucunda, eksik kayıtların büyük çoğunluğunun order_status değeri shipped olarak görülmüştür. Bunun yanında daha az sayıda canceled, unavailable, invoiced ve processing durumundaki siparişler de bulunmaktadır. Bu durum, eksik değerlerin yalnızca iptal edilen siparişlere özgü olmadığını; sevkiyat aşamasındaki veya henüz tamamlanmamış siparişlerde de görülebildiğini göstermektedir.
